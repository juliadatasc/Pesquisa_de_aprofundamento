## 1. Escolha da pesquisa

A pesquisa selecionada foi o **Sistema de Avaliação da Educação Básica (Saeb) - 2023**, conduzida pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (Inep). Essa avaliação mede a proficiência dos estudantes brasileiros em Língua Portuguesa e Matemática em diferentes etapas de ensino, sendo uma das principais fontes de dados sobre a qualidade da educação no país.


## 2. Características gerais e justificativa

O Saeb avalia amostras representativas de alunos do 5º e 9º anos do Ensino Fundamental e da 3ª série do Ensino Médio, considerando também fatores como rede administrativa (estadual, municipal, privada), localização da escola (urbana ou rural) e capital/interior.

A escolha dessa pesquisa se justifica pela sua relevância para o diagnóstico da aprendizagem e pela possibilidade de observar a evolução do desempenho educacional em Pernambuco, estado que tem se destacado nacionalmente por avanços na educação pública nos últimos anos. Além disso, o banco de dados do Saeb permite explorar tanto as médias de proficiência quanto a distribuição dos alunos por níveis de desempenho, o que possibilita uma análise mais detalhada.

## 3. Informações e série de dados comentada para Pernambuco

Foram selecionadas duas informações principais:

**a) Médias de proficiência em Língua Portuguesa e Matemática**

**b) Distribuição dos alunos por níveis de proficiência em Matemática (Ensino Médio)**

**a) Médias de Proficiência**
O gráfico de linha gerado apresenta as médias de proficiência em Língua Portuguesa e Matemática para o 5º ano, 9º ano e 3ª série do Ensino Médio em Pernambuco.

Os resultados mostram uma tendência de crescimento contínuo nas médias ao longo das etapas de ensino.

No 5º ano, as médias são mais baixas (cerca de 200 pontos em LP e 210 em MT).
No 9º ano, há um salto significativo, com médias próximas de 260 pontos.
Na 3ª série do Ensino Médio, as médias ultrapassam 280 pontos em ambas as disciplinas.

Esses resultados indicam melhora progressiva da proficiência, embora seja importante ressaltar que os níveis ainda estão abaixo das metas recomendadas pelo MEC para a plena aprendizagem.

**b) Distribuição dos alunos por níveis – Matemática (Ensino Médio)**
O segundo gráfico apresenta a distribuição percentual dos alunos pernambucanos por níveis de proficiência em Matemática no Ensino Médio.

Observa-se que a maior concentração de alunos está nos níveis intermediários (entre 2 e 5), o que sugere que muitos estudantes ainda não atingiram níveis avançados de domínio da disciplina. Há também uma quantidade significativa de alunos nos níveis iniciais (0 e 1), representando dificuldades de aprendizagem acumuladas ao longo da trajetória escolar.

Os níveis mais altos (8, 9 e 10) apresentam percentuais reduzidos, evidenciando a necessidade de reforço pedagógico e políticas públicas voltadas à aprendizagem em Matemática no Ensino Médio.

## 4. Conclusão

A análise dos dados do Saeb 2023 para Pernambuco mostra que, embora o estado apresente uma evolução positiva nas médias de proficiência ao longo das etapas de ensino, persistem desafios em Matemática, especialmente no Ensino Médio. A ampliação de políticas educacionais de acompanhamento e reforço nas disciplinas de base é essencial para reduzir as desigualdades e elevar o nível de aprendizagem dos estudantes pernambucanos.

#### Importação dos dados e bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("TS_UF_20250507.xlsx")
```

```python
print("Colunas disponíveis:", df.columns.tolist())
```

```python
df.head()
```

#### Filtragem e análise descritiva

```python
pe = df[df["NO_UF"] == "Pernambuco"]
```

#### Cálculo das médias e gráfico de proficiência

```python
pe.head(10)
```

```python
# Selecionar apenas colunas de médias
medias = pe[["NO_UF","DEPENDENCIA_ADM","LOCALIZACAO","CAPITAL",
             "MEDIA_5_LP","MEDIA_5_MT","MEDIA_9_LP","MEDIA_9_MT",
             "MEDIA_12_LP","MEDIA_12_MT"]]
```

```python
# Calcular médias gerais para Pernambuco
medias_pe = medias.mean(numeric_only=True)
print(medias_pe)
```

```python
# Gráfico comparando LP e MT no 5º, 9º e 12º ano
anos = ["5º Ano", "9º Ano", "3ª Série EM"]
lp = [medias_pe["MEDIA_5_LP"], medias_pe["MEDIA_9_LP"], medias_pe["MEDIA_12_LP"]]
mt = [medias_pe["MEDIA_5_MT"], medias_pe["MEDIA_9_MT"], medias_pe["MEDIA_12_MT"]]

plt.plot(anos, lp, marker="o", label="Língua Portuguesa")
plt.plot(anos, mt, marker="o", label="Matemática")

plt.title("Proficiência Média - Pernambuco (Saeb 2023)")
plt.xlabel("Etapa de Ensino")
plt.ylabel("Média de Proficiência")
plt.legend()
plt.grid(True)
plt.show()
```

Observa-se um crescimento contínuo nas médias de proficiência em ambas as disciplinas, indicando progressos na aprendizagem ao longo das etapas de ensino.
No entanto, as médias ainda se mantêm abaixo das metas nacionais de desempenho.

#### Distribuição por níveis – Matemática

```python
# Selecionar apenas colunas dos níveis de Matemática do Ensino Médio
niveis_mt = [col for col in pe.columns if "nivel" in col and "MTEM" in col]

dist_mt = pe[niveis_mt].mean()

# Plotar gráfico de barras
dist_mt.plot(kind="bar", figsize=(10,5))
plt.title("Distribuição dos alunos por níveis - Matemática (Ensino Médio, PE 2023)")
plt.xlabel("Nível de Proficiência")
plt.ylabel("Percentual (%)")
plt.show()
```

A maior parte dos alunos concentra-se nos níveis intermediários, com percentuais menores nos níveis mais altos, o que demonstra desafios na consolidação da aprendizagem em Matemática no Ensino Médio.
