## 1. Escolha da pesquisa

A pesquisa selecionada foi o **Sistema de Avaliação da Educação Básica (Saeb) - 2023**, conduzida pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (Inep). Essa avaliação mede a proficiência dos estudantes brasileiros em Língua Portuguesa e Matemática em diferentes etapas de ensino, sendo uma das principais fontes de dados sobre a qualidade da educação no país.


## 2. Características gerais e justificativa

O Saeb é uma das principais pesquisas realizadas pelo Inep e tem como objetivo avaliar a qualidade da educação básica no Brasil. Ele mede a proficiência dos estudantes em áreas fundamentais, como Língua Portuguesa e Matemática, em diferentes etapas de ensino (5º ano, 9º ano e 3ª série do Ensino Médio). A escolha dessa pesquisa se justifica pela sua relevância para o diagnóstico educacional — os dados permitem identificar avanços e desafios no desempenho dos estudantes de cada estado, subsidiando políticas públicas e ações voltadas à melhoria da aprendizagem. No caso de Pernambuco, os resultados oferecem um panorama valioso sobre o progresso do ensino no estado e sua comparação com a média nacional.

## 3. Informações e série de dados comentada para Pernambuco

A partir do conjunto de dados “TS_UF_20250507.xlsx”, foram calculadas as médias de proficiência de Pernambuco e do Brasil em Língua Portuguesa (LP) e Matemática (MT), considerando as três etapas de ensino avaliadas.

**Língua Portuguesa – Pernambuco:**

* 5º ano: ~200 pontos
* 9º ano: ~260 pontos
* 3ª série do EM: ~283 pontos

**Matemática – Pernambuco:**

* 5º ano: ~210 pontos
* 9º ano: ~259 pontos
* 3ª série do EM: ~285 pontos

Os resultados mostram **crescimento progressivo das médias** ao longo das etapas, indicando que o desempenho dos estudantes melhora conforme o avanço escolar. Em Pernambuco, o desempenho em Matemática é ligeiramente superior ao de Língua Portuguesa em quase todas as etapas.

Quando comparadas às **médias nacionais**, observa-se que Pernambuco apresenta **desempenhos próximos, porém um pouco abaixo da média do Brasil em algumas etapas**, especialmente no 5º ano. No entanto, a diferença tende a diminuir no Ensino Médio, revelando uma **trajetória de melhora contínua** e aproximação dos resultados nacionais.

## 4. Conclusão

A análise evidencia que Pernambuco apresenta avanços consistentes na aprendizagem, especialmente entre o 5º e o 9º ano, tanto em Língua Portuguesa quanto em Matemática. A convergência com as médias nacionais sugere que as políticas educacionais estaduais têm contribuído para reduzir as desigualdades de desempenho. Ainda assim, os resultados apontam a necessidade de fortalecer as aprendizagens nos anos iniciais, base para o sucesso nas etapas seguintes.

## Código completo

```python
# Importação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura dos dados
df = pd.read_excel("TS_UF_20250507.xlsx")
```

```python
# Verificação de colunas disponíveis
print("Colunas disponíveis:", df.columns.tolist())
df.head()
```

```python
# Cálculo das médias nacionais
medias_nac = df[
    ["MEDIA_5_LP", "MEDIA_5_MT",
     "MEDIA_9_LP", "MEDIA_9_MT",
     "MEDIA_12_LP", "MEDIA_12_MT"]
].mean(numeric_only=True)

print("\nMédias nacionais (Saeb 2023):")
print(medias_nac)
```

```python
# Filtragem dos dados para Pernambuco
pe = df[df["NO_UF"] == "Pernambuco"]
pe.head()
```

```python
# Seleção das colunas das médias
medias = pe[
    ["NO_UF", "DEPENDENCIA_ADM", "LOCALIZACAO", "CAPITAL",
     "MEDIA_5_LP", "MEDIA_5_MT",
     "MEDIA_9_LP", "MEDIA_9_MT",
     "MEDIA_12_LP", "MEDIA_12_MT"]
]
```

```python
# Cálculo das médias gerais para Pernambuco
medias_pe = medias.mean(numeric_only=True)
print("Médias - Pernambuco:")
print(medias_pe)
```

```python
# Gráfico 1: Proficiência média por etapa em Pernambuco
anos = ["5º Ano", "9º Ano", "3ª Série EM"]
lp = [medias_pe["MEDIA_5_LP"], medias_pe["MEDIA_9_LP"], medias_pe["MEDIA_12_LP"]]
mt = [medias_pe["MEDIA_5_MT"], medias_pe["MEDIA_9_MT"], medias_pe["MEDIA_12_MT"]]

plt.figure(figsize=(8,5))
plt.plot(anos, lp, marker="o", label="Língua Portuguesa")
plt.plot(anos, mt, marker="o", label="Matemática")

plt.title("Proficiência Média - Pernambuco (Saeb 2023)")
plt.xlabel("Etapa de Ensino")
plt.ylabel("Média de Proficiência")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico 2: Comparativo Pernambuco x Brasil
# Médias de Pernambuco
lp_pe = [medias_pe["MEDIA_5_LP"], medias_pe["MEDIA_9_LP"], medias_pe["MEDIA_12_LP"]]
mt_pe = [medias_pe["MEDIA_5_MT"], medias_pe["MEDIA_9_MT"], medias_pe["MEDIA_12_MT"]]

# Médias nacionais
lp_br = [medias_nac["MEDIA_5_LP"], medias_nac["MEDIA_9_LP"], medias_nac["MEDIA_12_LP"]]
mt_br = [medias_nac["MEDIA_5_MT"], medias_nac["MEDIA_9_MT"], medias_nac["MEDIA_12_MT"]]

plt.figure(figsize=(8,5))
plt.plot(anos, lp_pe, marker='o', label='LP - Pernambuco')
plt.plot(anos, mt_pe, marker='o', label='MT - Pernambuco')
plt.plot(anos, lp_br, marker='s', linestyle='--', label='LP - Brasil')
plt.plot(anos, mt_br, marker='s', linestyle='--', label='MT - Brasil')

plt.title("Comparativo de Proficiência Média – Pernambuco x Brasil (Saeb 2023)")
plt.xlabel("Etapa de Ensino")
plt.ylabel("Média de Proficiência")
plt.legend()
plt.grid(True)
plt.show()
```
